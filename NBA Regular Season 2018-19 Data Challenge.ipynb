```python
nbastat = pd.read_csv("stats.csv", sep = ',')
nbastat
```

# What was the average age of player in the league?

```python
nbastat.mean(numeric_only= True)[['Age']]
```

# What player scored the most points?

```python
nbastat[nbastat.Points == nbastat.Points.max()][['Name','Points']]
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
nbastat[nbastat.Blocks == nbastat.Blocks.max()][['Name','Blocks']]
# Yes it was a post player.
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
nbastat.groupby(['Team']).mean([['Points']]).sort_values('Points', ascending = False)[['Points']].head()
# There was not a win % column so I just decided to take the averages of all the points and order them. So if the Bucks consistently
# scored the most points they probably had the highest chance of winning because no one is scoring more than them.
```

# What player had the best 3-pt percentage? 

```python
nbastat.groupby(['Team']).max()[['Name','FG3%']].head()
```

# Who played the most minutes during the season

```python
nbastat[nbastat.MP == nbastat.MP.max()][['Name','MP']]
```

# What player given their player effiecency rating was the clutchest during the season?

```python
nbastat.groupby('Name').max(numeric_only= True).sort_values('PER', ascending = False)[["PER"]].head()
```

# What team had the youngest roster?

```python
nbastat.groupby('Team').mean(numeric_only= True).sort_values('Age', ascending = True)[["Age"]].head()
```

# Who is the highest paid player during the seasion?

```python
#nbastat['Salary'] = pd.to_numeric(nbastat['Salary'])
#nbastat.Salary.
nbastat['Salary'] = nbastat['Salary'].replace('-',pd.NA).fillna(0).astype(int)
nbastat.groupby('Name').mean(numeric_only= True).sort_values('Salary', ascending = False)[["Salary"]].head()
```

# At the end of a game, who WOULDN'T you want on the Free Throw Line?

```python
nbastat.groupby('Name').min(numeric_only= True).sort_values('FT%', ascending = True)[["FT%"]].head()
```
